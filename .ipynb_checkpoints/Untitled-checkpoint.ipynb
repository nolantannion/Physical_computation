```python
# -*- coding: utf-8 -*-

from math import pi, sqrt
import numpy as np
from numpy import zeros
from vpython import sphere, rate, color, canvas, vector

# Parámetros del sistema (unidades: AU, años)
GMSol = 4 * pi**2

T = 76                # periodo del cometa Halley (años)
dt = 0.001            # paso temporal (años)
tiempoTotal = 1 * T   # tiempo total de simulación

N = int(tiempoTotal / dt)

# Vectores
x = zeros(N)
y = zeros(N)
r = zeros(N)
vx = zeros(N)
vy = zeros(N)
t = zeros(N)

# Condiciones iniciales (perihelio del Halley)
t[0] = 0
x[0] = 0.59
y[0] = 0

r[0] = sqrt(x[0]**2 + y[0]**2)

vx[0] = 0
vy[0] = 11.5

# Integración Euler-Cromer
for i in range(N-1):

    ax = -GMSol * x[i] / r[i]**3
    ay = -GMSol * y[i] / r[i]**3

    vx[i+1] = vx[i] + ax * dt
    vy[i+1] = vy[i] + ay * dt

    x[i+1] = x[i] + vx[i+1] * dt
    y[i+1] = y[i] + vy[i+1] * dt

    r[i+1] = sqrt(x[i+1]**2 + y[i+1]**2)
    t[i+1] = t[i] + dt

r = np.sqrt(x**2 + y**2)
print(np.max(r), x[-1], y[-1])

# Visualización
canvas(title="Órbita del cometa Halley",
       width=700, height=700,
       center=vector(0,0,0),
       background=vector(0,0,0))

sol = sphere(pos=vector(0,0,0),
             color=color.yellow,
             radius=0.3)

planeta = sphere(pos=vector(x[0],y[0],0),
                 color=color.cyan,
                 radius=0.08)

# Animación
for i in range(N):

    rate(100)

    px = x[i]
    py = y[i]

    sphere(pos=vector(px,py,0),
           color=color.red,
           radius=0.01)

    planeta.pos = vector(px,py,0)
```
